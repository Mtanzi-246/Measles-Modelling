```python
# calling the libraries
import numpy as np # for arrays(vectors)
import scipy.integrate as sci #(solving)
#import matplotlib.pyplot as plt #( for plotting)
import matplotlib.pyplot as plt; #plt.rcdefaults()
import pylab
import matplotlib.lines
from matplotlib import lines
```

```python
 # function
def measles(x, T):
    """This is a function that stores the state variables of the system"""
    S = x[0]   # Susceptible
    I = x[1] # Infected
    R = x[2]   # Recovered

    # Parameter values
    deltaT = 1/365 # The function to change the parameters to days instead of years
    m = 0.02*deltaT
    beta = 200*deltaT
    gamma = 50*deltaT


    Y = np.zeros(3)

    # The differential equations where (X and Y) are constants
    Y[0] = m*(S+I+R) - (beta*I + m)*S
    Y[1] = beta*I*S - (gamma + m)*I
    Y[2] = gamma*I - m*R

    return Y

t_start = 0    # start time
t_end = 365    # end time
t_step = 365  # time steps
T = np.linspace(t_start, t_end, t_step)

# Initial values of the compartments
sol0 = np.array([0.9, 0.1, 0]) # initial conditions by proportions don't be an idiot.

# Solving the ordinary equation using the scipy function odeintegrate odeint()

sol1 = sci.odeint(measles, sol0, T)
solution_final = sol1.T

print(sol1)
# print(solution_final)
```

```python
plt.figure(figsize=(11,8))
plt.plot(T, solution_final[0], '-.', label ='S', linewidth=3, color = 'red')
plt.plot(T, solution_final[1], label ='I', linewidth=3)
plt.plot(T, solution_final[2], label ='R', linewidth=3)
pylab.ylabel('Population proportions', fontsize=14, color='blue')
pylab.xlabel('Time', fontsize=14)
plt.title('Measles Model Simulation', fontsize=16, color = 'red')
plt.legend(bbox_to_anchor = [1,0.5])
plt.show()
```
